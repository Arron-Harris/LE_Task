 Help Acme Corp to help make a data-driven decision about what compensation scheme to implement in new plants. The spreadsheet came from the manager of one of Acme’s existing plants. The manager that sent it provided the following details:


•	The data cover one year (50 weeks of operation) at the plant.

•	This plant produces only one product – electrodes.

•	The plant was opened at the beginning of the year (week 1).

•	Each person’s individual electrode output is measured separately. “Output” (column C of the “All Data” tab of the spreadsheet) is the number of electrodes the person produced during the week.

•	Initially, workers were paid a fixed salary equal to 25 currency units per hour. The standard work week is forty hours and there is no overtime work, so all workers had pay of 1000 units per week.

•	In the second half of the year (starting in week 26), compensation switched to a piece rate with a guaranteed minimum of 875 currency units per week. Each employee was paid the greater of 875 or 1*(electrode output) for the week. So, if a person produced 1100 electrodes, he/she earned 1100 (because this is greater than 875). If a person produced 800 electrodes, he/she earned 875 (because 1*800=800 is less than the guaranteed minimum of 875)

•	The “All Data” tab of the spreadsheet shows the worker, the week, electrode output, and pay for every worker-week during the year.

1.Did the piece rate motivate workers?

Here I would check the output column and the piece rate column.

If the piece rate did motivate the workers then we will see a higher output


```python
#workers who have not participated in the piece rate system
piece_rate0 = LE[LE['PieceRate']== 0]
```

```python
#workers who have participated in the piece rate system
piece_rate1 = LE[LE['PieceRate']== 1]
```

```python
piece_rate0.head()
```

```python
piece_rate1
```

```python
#Here we see the output of people who are on the piece rate system
sns.swarmplot(x ='PieceRate',y = 'Output', data= piece_rate1 )
```

```python
##Here we see the output of people who are not on the piece rate system
sns.swarmplot(x ='PieceRate',y = 'Output', data= piece_rate0 )
```

Yes the piece rate did motivate some workers.

We see a jump of output from 900 all the way to 1100.

Without the piece rate it stayed at 900.
